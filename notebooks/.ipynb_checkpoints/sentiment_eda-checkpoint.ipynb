# Sentiment Analysis EDA

Exploratory Data Analysis for sentiment analysis results:
- Overall sentiment distribution
- Sentiment by bank and rating
- Sentiment score distributions
- Correlation between ratings and sentiment


## Setup and Imports


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Add project root to path
PROJECT_ROOT = Path().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from config import (
    PROCESSED_DATA_PATH,
    SENTIMENT_DATA_PATH,
    SENTIMENT_SUMMARY_PATH,
    BANKS,
)

# Visualization libraries (optional)
try:
    import matplotlib.pyplot as plt
    import seaborn as sns
    HAS_VIZ = True
    sns.set_style("whitegrid")
    plt.rcParams['figure.figsize'] = (12, 6)
except ImportError:
    HAS_VIZ = False
    print("Note: matplotlib/seaborn not available. Visualizations will be skipped.")
```

```python
# Load sentiment-enriched data
if not SENTIMENT_DATA_PATH.exists():
    print("="*60)
    print("ERROR: Sentiment data file not found!")
    print("="*60)
    print(f"Expected file: {SENTIMENT_DATA_PATH}")
    print("\nPlease run the sentiment analysis script first:")
    print("  python scripts/sentiment_analysis.py")
    print("\nThis will generate the required sentiment data file.")
    raise FileNotFoundError(f"Sentiment data file not found: {SENTIMENT_DATA_PATH}")

df_sentiment = pd.read_csv(SENTIMENT_DATA_PATH)
print(f"Loaded {len(df_sentiment):,} reviews with sentiment scores")
print(f"\nSentiment engine used: {df_sentiment['sentiment_source'].iloc[0] if 'sentiment_source' in df_sentiment.columns else 'N/A'}")
print("\n" + "="*60)
print("Sentiment Distribution Overview")
print("="*60)
sentiment_counts = df_sentiment['sentiment_label'].value_counts()
for label, count in sentiment_counts.items():
    pct = (count / len(df_sentiment)) * 100
    print(f"{label.capitalize()}: {count:,} ({pct:.1f}%)")

if HAS_VIZ:
    plt.figure(figsize=(10, 5))
    sentiment_counts.plot(kind='bar', color=['green', 'gray', 'red'])
    plt.title('Overall Sentiment Distribution', fontsize=14, fontweight='bold')
    plt.xlabel('Sentiment Label', fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# Sentiment by bank
print("\n" + "="*60)
print("Sentiment Distribution by Bank")
print("="*60)
sentiment_by_bank = pd.crosstab(df_sentiment['bank'], df_sentiment['sentiment_label'])
sentiment_by_bank_pct = sentiment_by_bank.div(sentiment_by_bank.sum(axis=1), axis=0) * 100

for bank in sentiment_by_bank.index:
    print(f"\n{bank}:")
    for label in ['positive', 'neutral', 'negative']:
        if label in sentiment_by_bank.columns:
            count = sentiment_by_bank.loc[bank, label]
            pct = sentiment_by_bank_pct.loc[bank, label]
            print(f"  {label.capitalize()}: {count:,} ({pct:.1f}%)")

if HAS_VIZ:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Stacked bar chart
    sentiment_by_bank.plot(kind='bar', stacked=True, ax=axes[0], 
                          color=['green', 'gray', 'red'])
    axes[0].set_title('Sentiment Distribution by Bank (Counts)', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Bank', fontsize=12)
    axes[0].set_ylabel('Count', fontsize=12)
    axes[0].legend(title='Sentiment')
    axes[0].tick_params(axis='x', rotation=45)
    
    # Percentage stacked bar
    sentiment_by_bank_pct.plot(kind='bar', stacked=True, ax=axes[1],
                               color=['green', 'gray', 'red'])
    axes[1].set_title('Sentiment Distribution by Bank (Percentages)', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Bank', fontsize=12)
    axes[1].set_ylabel('Percentage', fontsize=12)
    axes[1].legend(title='Sentiment')
    axes[1].tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Sentiment by rating
print("\n" + "="*60)
print("Sentiment vs Rating Analysis")
print("="*60)
sentiment_by_rating = pd.crosstab(df_sentiment['rating'], df_sentiment['sentiment_label'])

for rating in sorted(sentiment_by_rating.index):
    print(f"\n{'' * int(rating)} Rating ({rating}):")
    total = sentiment_by_rating.loc[rating].sum()
    for label in ['positive', 'neutral', 'negative']:
        if label in sentiment_by_rating.columns:
            count = sentiment_by_rating.loc[rating, label]
            pct = (count / total) * 100 if total > 0 else 0
            print(f"  {label.capitalize()}: {count:,} ({pct:.1f}%)")

if HAS_VIZ:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Sentiment by rating
    sentiment_by_rating.plot(kind='bar', ax=axes[0], color=['green', 'gray', 'red'])
    axes[0].set_title('Sentiment Distribution by Rating', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Rating', fontsize=12)
    axes[0].set_ylabel('Count', fontsize=12)
    axes[0].legend(title='Sentiment')
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
    
    # Average sentiment score by rating
    avg_score_by_rating = df_sentiment.groupby('rating')['sentiment_score'].mean()
    avg_score_by_rating.plot(kind='line', ax=axes[1], marker='o', linewidth=2, markersize=8)
    axes[1].set_title('Average Sentiment Score by Rating', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Rating', fontsize=12)
    axes[1].set_ylabel('Average Sentiment Score', fontsize=12)
    axes[1].grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
# Load and display sentiment summary
if SENTIMENT_SUMMARY_PATH.exists():
    df_sentiment_summary = pd.read_csv(SENTIMENT_SUMMARY_PATH)
    print("\n" + "="*60)
    print("Sentiment Summary by Bank and Rating")
    print("="*60)
    print(df_sentiment_summary.to_string(index=False))
    
    if HAS_VIZ:
        # Heatmap of positive sentiment percentage
        pivot_positive = df_sentiment_summary.pivot(index='bank', columns='rating', values='positive_pct')
        plt.figure(figsize=(10, 6))
        sns.heatmap(pivot_positive, annot=True, fmt='.1f', cmap='RdYlGn', 
                   cbar_kws={'label': 'Positive Sentiment %'})
        plt.title('Positive Sentiment Percentage by Bank and Rating', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
```

```python
# Sentiment score distribution
print("\n" + "="*60)
print("Sentiment Score Statistics")
print("="*60)
print(f"Mean sentiment score: {df_sentiment['sentiment_score'].mean():.3f}")
print(f"Median sentiment score: {df_sentiment['sentiment_score'].median():.3f}")
print(f"Min sentiment score: {df_sentiment['sentiment_score'].min():.3f}")
print(f"Max sentiment score: {df_sentiment['sentiment_score'].max():.3f}")

print("\nSentiment score by label:")
for label in df_sentiment['sentiment_label'].unique():
    label_scores = df_sentiment[df_sentiment['sentiment_label'] == label]['sentiment_score']
    print(f"  {label.capitalize()}: mean={label_scores.mean():.3f}, median={label_scores.median():.3f}")

if HAS_VIZ:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histogram of sentiment scores
    df_sentiment['sentiment_score'].hist(bins=30, ax=axes[0], color='steelblue', edgecolor='black')
    axes[0].set_title('Distribution of Sentiment Scores', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Sentiment Score', fontsize=12)
    axes[0].set_ylabel('Frequency', fontsize=12)
    
    # Box plot by sentiment label
    df_sentiment.boxplot(column='sentiment_score', by='sentiment_label', ax=axes[1])
    axes[1].set_title('Sentiment Score Distribution by Label', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Sentiment Label', fontsize=12)
    axes[1].set_ylabel('Sentiment Score', fontsize=12)
    plt.suptitle('')
    plt.tight_layout()
    plt.show()
else:
    print("\nNote: Install matplotlib and seaborn to see visualizations.")
```

```python
# Coverage analysis
df_clean = pd.read_csv(PROCESSED_DATA_PATH)
sentiment_coverage = (len(df_sentiment[df_sentiment['sentiment_label'].notna()]) / len(df_clean)) * 100
print("\n" + "="*60)
print("Sentiment Analysis Coverage")
print("="*60)
print(f"Total preprocessed reviews: {len(df_clean):,}")
print(f"Reviews with sentiment scores: {len(df_sentiment):,}")
print(f"Coverage: {sentiment_coverage:.1f}%")
```
