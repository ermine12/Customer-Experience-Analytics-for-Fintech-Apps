# Preprocessing EDA

Exploratory Data Analysis for preprocessed reviews:
- Dataset overview and quality metrics
- Reviews distribution by bank
- Rating distribution analysis
- Temporal trends (if available)


## Setup and Imports


```python
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Add project root to path
PROJECT_ROOT = Path().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from config import PROCESSED_DATA_PATH, BANKS

# Visualization libraries (optional)
try:
    import matplotlib.pyplot as plt
    import seaborn as sns
    HAS_VIZ = True
    sns.set_style("whitegrid")
    plt.rcParams['figure.figsize'] = (12, 6)
except ImportError:
    HAS_VIZ = False
    print("Note: matplotlib/seaborn not available. Visualizations will be skipped.")
```

```python
# Load preprocessed data
if not PROCESSED_DATA_PATH.exists():
    print("="*60)
    print("ERROR: Preprocessed data file not found!")
    print("="*60)
    print(f"Expected file: {PROCESSED_DATA_PATH}")
    print("\nPlease run the preprocessing script first:")
    print("  python scripts/preprocess_reviews.py")
    print("\nThis will generate the required preprocessed data file.")
    raise FileNotFoundError(f"Preprocessed data file not found: {PROCESSED_DATA_PATH}")

df_clean = pd.read_csv(PROCESSED_DATA_PATH)
print(f"Loaded {len(df_clean):,} preprocessed reviews")
print(f"Columns: {', '.join(df_clean.columns)}")
print("\n" + "="*60)
print("Dataset Overview")
print("="*60)
df_clean.info()
```

```python
# Basic statistics
print("\n" + "="*60)
print("Basic Statistics")
print("="*60)
print(f"Total reviews: {len(df_clean):,}")
print(f"Unique banks: {df_clean['bank'].nunique()}")
print(f"Date range: {df_clean['date'].min()} to {df_clean['date'].max()}")

if 'review_length' in df_clean.columns:
    print(f"\nReview length statistics:")
    print(f"  Mean: {df_clean['review_length'].mean():.1f} characters")
    print(f"  Median: {df_clean['review_length'].median():.1f} characters")
    print(f"  Min: {df_clean['review_length'].min()}")
    print(f"  Max: {df_clean['review_length'].max()}")
```

```python
# Reviews per bank
print("\n" + "="*60)
print("Reviews per Bank")
print("="*60)
bank_counts = df_clean['bank'].value_counts()
for bank, count in bank_counts.items():
    pct = (count / len(df_clean)) * 100
    print(f"{bank}: {count:,} ({pct:.1f}%)")

if HAS_VIZ:
    plt.figure(figsize=(10, 5))
    bank_counts.plot(kind='bar', color='steelblue')
    plt.title('Number of Reviews per Bank', fontsize=14, fontweight='bold')
    plt.xlabel('Bank', fontsize=12)
    plt.ylabel('Number of Reviews', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Rating distribution
print("\n" + "="*60)
print("Rating Distribution")
print("="*60)
rating_counts = df_clean['rating'].value_counts().sort_index(ascending=False)
for rating, count in rating_counts.items():
    pct = (count / len(df_clean)) * 100
    stars = '⭐' * int(rating)
    print(f"{stars} ({rating}): {count:,} ({pct:.1f}%)")

if HAS_VIZ:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Overall rating distribution
    rating_counts.sort_index().plot(kind='bar', ax=axes[0], color='coral')
    axes[0].set_title('Overall Rating Distribution', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Rating', fontsize=12)
    axes[0].set_ylabel('Count', fontsize=12)
    axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=0)
    
    # Rating distribution by bank
    rating_by_bank = pd.crosstab(df_clean['bank'], df_clean['rating'])
    rating_by_bank.plot(kind='bar', stacked=True, ax=axes[1], colormap='viridis')
    axes[1].set_title('Rating Distribution by Bank', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Bank', fontsize=12)
    axes[1].set_ylabel('Count', fontsize=12)
    axes[1].legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
    axes[1].tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Temporal analysis
print("\n" + "="*60)
print("Temporal Distribution")
print("="*60)

# Parse dates
df_clean['date_parsed'] = pd.to_datetime(df_clean['date'], errors='coerce')
df_clean['review_year'] = df_clean['date_parsed'].dt.year
df_clean['review_month'] = df_clean['date_parsed'].dt.month

# Reviews by year
year_counts = df_clean['review_year'].value_counts().sort_index()
print("\nReviews by Year:")
for year, count in year_counts.items():
    if pd.notna(year):
        print(f"  {int(year)}: {count:,}")

if HAS_VIZ:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Reviews over time
    monthly = df_clean.groupby([df_clean['date_parsed'].dt.to_period('M')]).size()
    monthly.plot(kind='line', ax=axes[0], marker='o')
    axes[0].set_title('Reviews Over Time (Monthly)', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Month', fontsize=12)
    axes[0].set_ylabel('Number of Reviews', fontsize=12)
    axes[0].tick_params(axis='x', rotation=45)
    
    # Reviews by bank over time
    for bank in df_clean['bank'].unique():
        bank_data = df_clean[df_clean['bank'] == bank]
        monthly_bank = bank_data.groupby([bank_data['date_parsed'].dt.to_period('M')]).size()
        axes[1].plot(monthly_bank.index.astype(str), monthly_bank.values, marker='o', label=bank)
    axes[1].set_title('Reviews Over Time by Bank', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Month', fontsize=12)
    axes[1].set_ylabel('Number of Reviews', fontsize=12)
    axes[1].legend()
    axes[1].tick_params(axis='x', rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("\nNote: Install matplotlib and seaborn to see temporal visualizations.")
```

```python
# Sample reviews by rating
print("\n" + "="*60)
print("Sample Reviews by Rating")
print("="*60)
for rating in sorted(df_clean['rating'].unique(), reverse=True):
    sample = df_clean[df_clean['rating'] == rating].head(2)
    if not sample.empty:
        print(f"\n{'⭐' * int(rating)} Rating ({rating}):")
        for idx, row in sample.iterrows():
            print(f"  Bank: {row['bank']}")
            print(f"  Review: {row['review'][:200]}...")
            print()
```
